```python
movie_np = []
# Iterate over the DataFrame rows
for cust_id in df['Cust_id']:
    if ':' in cust_id:
        # Update the current movie ID
        movie_id = int(cust_id.replace(':',''))
    movie_np.append(movie_id)
```

```python
movie_np
```

```python
df['Movie_id'] = movie_np
df.head()
```

```python
df = df[df['Rating'].notna()]
df
```

```python
df.shape
```

```python
df.info()
```

```python
df['Cust_id'] = df['Cust_id'].astype(int)
df.dtypes
```

```python
'''
pre-filtering
now we will remove all the users that have rated less movies and
also all those movies that has been rated less in numbers
'''
```

```python
# Remove all those movies that has been rated less in numbers
movie_summary = df.groupby('Movie_id')['Rating'].agg(['count'])
movie_summary
```

```python
# now we will create a benchmark
movie_benchmark = round(movie_summary['count'].quantile(0.6),0)
movie_benchmark
```

```python
drop_movie_list = movie_summary[movie_summary['count'] < movie_benchmark].index
drop_movie_list
```

```python
len(drop_movie_list)
```

```python
# Movies left
4499-2699
```

```python
# now we will remove all the users that are in-active
cust_summary = df.groupby('Cust_id')['Rating'].agg(['count'])
cust_summary
```

```python
cust_benchmark = round(cust_summary['count'].quantile(0.6),0)
cust_benchmark
```

```python
drop_cust_list = cust_summary[cust_summary['count'] < cust_benchmark].index
drop_cust_list
```

```python
len(drop_cust_list)
```

```python
df = df[~df['Movie_id'].isin(drop_movie_list)]
df = df[~df['Cust_id'].isin(drop_cust_list)]
print('after trimming the data, the shape is: {}'.format(df.shape))
```
